### Task 4: Create a new DataFrame containing only the "Bundesländer" (hint: use the corresponding SATZART key)

```python
df = pd.read_excel("demo_excel.xls", sheetname = 2, header=3, index_col=2)
df_bundesländer = df[df.ix[:,"SATZART"]==10]
df_bundesländer
```

### Task 5: Plot the age composition of the different Bundesländer

```python
age_composition = df_bundesländer.iloc[:,7:12]
age_composition.plot(kind="barh", stacked=True)
```

### Task 6: ...and now relative to the population in each Bundesland (hint the .div method of a DataFrame can be used to divide two dataframes)

```python
df_relative_age = df_bundesländer.ix[:,7:12].div(df_bundesländer.iloc[:,3], axis="index")
df_relative_age
df_relative_age.plot(kind="barh", stacked=True)
```

### Task 9: Merge the two data sets (hint: "Kreisschlüssel" is the same as "AGS")

```python
germs
df_merge = pd.merge(df, germs, left_on="AGS", right_on="Kreisschlussel")
df_merge.drop("Kreisschlussel", axis=1, inplace=True)
```

###  Task 10: Plot number of hospitalized patients per inhabitant against relative number of old people (65+)

```python
df_merge_drop = df_merge.drop(df_merge.columns[15:28], axis=1)
index_list = [7,8,9,10,11,14]
df_merge_drop
df_merge_drop_relative = df_merge_drop.ix[:,index_list].div(df_merge_drop.iloc[:,2], axis="index")
cities = df_merge.iloc[:,13:14]
df_merge_drop_relative_new = pd.concat([cities, df_merge_drop_relative], axis = 1)
df_merge_drop_relative_new
hospitalized_vs_old = df_merge_drop_relative_new.iloc[:,[0,5,6]]
hospitalized_vs_old
hospitalized_vs_old.set_index("Landkreis/Kreisfreie Stadt", inplace = True)
hospitalized_vs_old.plot(x = "ALTER_5",y = "Krankenhaus- Patienten 2013", kind="scatter")
```
